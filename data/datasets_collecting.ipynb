```python
YandexGPT1_df = pd.read_csv('generated_YandexGPT.csv')
YandexGPT2_df = pd.read_csv('generated_YaGPT2.csv')
YandexGPT3_df = pd.read_csv('generated_YaGPT3_05.csv')
YandexGPT4_df = pd.read_csv('generated_YaGPT4_05.csv')

YandexGPT_df = pd.concat([YandexGPT1_df,YandexGPT2_df,YandexGPT3_df,YandexGPT4_df])
YandexGPT_df
```

```python
YandexGPT_df=YandexGPT_df.drop(YandexGPT_df[YandexGPT_df.generated_text == 'К сожалению, я не могу ничего сказать об этом. Давайте сменим тему?'].index)    
YandexGPT_df=YandexGPT_df.drop(['Unnamed: 0'], axis=1)
YandexGPT_df = YandexGPT_df.drop_duplicates().reset_index(drop=True)
YandexGPT_df
```

```python
YandexGPT_df['generated'] = 1
YandexGPT_df['model'] = 'YandexGPT'
YandexGPT_df
```

```python
YandexGPT_df.to_csv('YandexGPT.csv')
```

```python
GigaChat_1_df = pd.read_csv('generated_GigaChat.csv')
GigaChat_2_df = pd.read_csv('generated_GigaChat2.csv')
GigaChat_3_df = pd.read_csv('generated_GigaChat3.csv')

GigaChat_df = pd.concat([GigaChat_1_df,GigaChat_2_df,GigaChat_3_df])
GigaChat_df
```

```python
GigaChat_df=GigaChat_df.drop(GigaChat_df[GigaChat_df.generated_text.str.find('Не люблю менять тему разговора, но вот сейчас')*GigaChat_df.generated_text.str.find('Что-то в вашем вопросе меня смущает') != 1].index)    
GigaChat_df=GigaChat_df.drop(GigaChat_df[GigaChat_df.generated_text == 'Как у нейросетевой языковой модели у меня не может быть настроения, но почему-то я совсем не хочу говорить на эту тему.'].index)    
GigaChat_df=GigaChat_df.drop(['Unnamed: 0'], axis=1)
GigaChat_df = GigaChat_df.drop_duplicates().reset_index(drop=True)
GigaChat_df
```

```python
GigaChat_df['generated'] = 1
GigaChat_df['model'] = 'GigaChat'
GigaChat_df
```
